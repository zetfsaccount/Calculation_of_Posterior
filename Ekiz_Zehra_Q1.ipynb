Let us numerically calculate the posterior (1). Suppose the stimulus distribution
p(s) is Gaussian with mean 20 and standard deviation 4. The measurement distribution p(x|s)
is Gaussian with standard deviation Ïƒ = 5. A Bayesian observer infers s from an observed
measurement xobs = 30. We are now going to calculate the posterior pdf using numerical methods.

a) Define a vector of hypothesized stimulus values s: [0, 0.2, 0.4, . . . , 40].

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

positionNormal = np.arange(0, 40.2, 0.2)
```

b) Compute the likelihood function and the prior on this vector of s values.

```python
mu1 = 20
sigma1 = 4
xobs = 30
sigma2 = 5

prior = norm.pdf(positionNormal,loc=mu1,scale=sigma1)
likelihood = norm.pdf(positionNormal,loc=xobs,scale=sigma2)
```

c) Multiply the likelihood and the prior (pointwise).

```python
multiplied = prior * likelihood
```

d) Divide this product by its sum over all s (normalization step).

```python
posterior_pmf = multiplied/sum(multiplied)
```

e) Convert this posterior probability mass function into a probability density function by dividing
by the step size you used in your vector of s values (e.g., 0.2).

```python
posterior = posterior_pmf/0.2
```

f) Plot the likelihood, prior, and posterior in the same plot.

```python
plt.plot(positionNormal, prior, 
         color = 'red', 
         linewidth = 2, label="Prior")

plt.plot(positionNormal, likelihood, 
         color = 'blue', 
         linewidth = 2, label="Likelihood")

plt.plot(positionNormal, posterior, 
         color = 'green', 
         linewidth = 2, label="Posterior")

plt.xlabel('s')
plt.ylabel('Probability density')
plt.title('Probability density function (pdf)')
plt.legend()
plt.show()
```

g) Is the posterior wider or narrower than likelihood and prior? Do you expect this based on the
equations we discussed?

Our posterior is narrower than our prior and likelihood. We expect that result because our posterior is more certain than our likelihood and prior so that it has less variance since we multiplied prior and likelihood.

h) Change the standard deviation of the measurement distribution to a large value. What happens
to the posterior? Can you explain this?

```python
mu1 = 20
sigma1 = 4
xobs = 30
sigma2 = 10


prior = norm.pdf(positionNormal,loc=mu1,scale=sigma1)
likelihood = norm.pdf(positionNormal,loc=xobs,scale=sigma2)

multiplied = prior * likelihood

posterior_pmf = multiplied/sum(multiplied)

posterior = posterior_pmf/0.2


plt.plot(positionNormal, prior, 
         color = 'red', 
         linewidth = 2, label="Prior")

plt.plot(positionNormal, likelihood, 
         color = 'blue', 
         linewidth = 2, label="Likelihood")

plt.plot(positionNormal, posterior, 
         color = 'green', 
         linewidth = 2, label="Posterior")

plt.xlabel('s')
plt.ylabel('Probability density')
plt.title('Probability density function (pdf)')
plt.legend()
plt.show()
```

Posterior becomes more affected by the prior. Since variance of our likelihood is very large that means it is uncertain, our prior determines our posterior.

i) Change the standard deviation of the measurement distribution to a small value. What
happens to the posterior? Can you explain this?

```python
mu1 = 20
sigma1 = 4
xobs = 30
sigma2 = 1


prior = norm.pdf(positionNormal,loc=mu1,scale=sigma1)
likelihood = norm.pdf(positionNormal,loc=xobs,scale=sigma2)

multiplied = prior * likelihood

posterior_pmf = multiplied/sum(multiplied)

posterior = posterior_pmf/0.2


plt.plot(positionNormal, prior, 
         color = 'red', 
         linewidth = 2, label="Prior")

plt.plot(positionNormal, likelihood, 
         color = 'blue', 
         linewidth = 2, label="Likelihood")

plt.plot(positionNormal, posterior, 
         color = 'green', 
         linewidth = 2, label="Posterior")

plt.xlabel('s')
plt.ylabel('Probability density, p(s)')
plt.title('Probability density function (pdf)')
plt.legend()
plt.show()
```

Posterior becomes more affected by the likelihood. Since variance of our likelihood is very small that means it is much certain than the prior, our likelihood determines our posterior.
